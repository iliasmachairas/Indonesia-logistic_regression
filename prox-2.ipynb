```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
holland=np.loadtxt('holland_temperature.dat')
seawater=np.loadtxt('holland_seawater.dat')
print(holland)
plt.figure(figsize=(10,5))
x=np.linspace(1,12,12)
print(x)
plt.plot(x,holland,'rs',)
plt.plot(x,seawater,linewidth=6)
plt.legend(["T","Seawater"],loc='best', fontsize='x-large')
plt.xlim(x.min()+0.5,x.max()+0.5)
plt.ylim(-2,25)
plt.xticks(np.linspace(4,10,6))
plt.yticks([0,10,12,26])
```

```python
Temp=np.loadtxt('holland_temperature.dat')
seawater=np.loadtxt('holland_seawater.dat')
x=np.linspace(1,12,12)
fig1, (ax1,ax2)=plt.subplots(2,1)
ax1.plot(x,Temp)
plt.xlabel("Chart-1-x-axis")
plt.ylabel("Chart-1-y-axis")
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Sep','Oct','Nov','Dec']
plt.xticks(np.arange(1,12),months)
ax2.plot(x,Temp-seawater)
plt.legend(["T"])
```

```python
Temp=np.loadtxt('holland_temperature.dat')
seawater=np.loadtxt('holland_seawater.dat')
x=np.linspace(1,12,12)
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(x,Temp)
plt.plot(x,seawater)
plt.xlabel("months")
plt.ylabel("Chart-1-y-axis")
plt.xticks(x,['Jan','Feb','Mar','Apr','May','Jun','Jul','Sep','Oct','Nov','Dec'])
plt.legend(["Temperature","Seawater"],fontsize='xx-large')
plt.subplot(2,1,2)
plt.plot(x,Temp-seawater)
plt.xlabel("months")
plt.ylabel("y-axis")
plt.legend(["difference"],fontsize='xx-large')
plt.xticks([1,2,3],['May','Jun','Jul'])
plt.yticks(np.linspace(10,15,2))
plt.savefig('foo.png')
```

```python
plt.figure(figsize=(10,10))
plt.fill([0,2,2,0,0],[0,0,2,2,0],'r',label='square')
plt.fill([1,2,2,1],[1,2,10,1],label='triangle')
R=1
Cx=0
Cy=4
u=np.linspace(0,2*np.pi,200)
x=Cx+R*np.cos(u)
y=Cy+R*np.sin(u)
plt.fill(x,y,label="circle",color='green')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.legend(fontsize=34)
plt.title("Scematic",fontsize=40)
```
